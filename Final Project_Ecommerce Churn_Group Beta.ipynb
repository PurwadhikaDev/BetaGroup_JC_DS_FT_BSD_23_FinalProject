## **1. Business Problem Understanding**

**Defining Customer Churn**

Customer churn refers to the number of customers that have stopped using company's products or services. In the context of e-commerce, customer churn refers specifically to the customers who have ceased making purchases or interacting with the online platform.

**Backgound**

A certain e-commerce company wants to predict whether its customers will churn or not in the future. With accurate predictions, the e-commerce company can take appropriate actions to retain customers if they are predicted not to continue shopping on the e-commerce platform (churn).

**Problems Statement**

In the midst of a highly competitive market, an e-commerce company needs to mitigate the risk of losing customers (customer churn). Customer shopping activities constitute the main source of revenue for e-commerce companies. However, the company lacks a system to filter or predict whether a customer will churn or not. This leads to many customers unexpectedly moving to other e-commerce companies (churn), necessitating the company to engage in customer acquisition efforts to replace those who churn. Generally, the cost of customer acquisition is much higher than the cost of customer retention.

Therefore, it would be beneficial for the e-commerce company to mitigate the risk of customer churn. The first step is to have a metric or model that can predict whether a customer will churn or not. With the predictions in hand, the e-commerce company can take appropriate actions targeted at the right customers. This allows for more effective and efficient allocation of budget and resources towards customer retention efforts, thereby generating outcomes that align with the company's expectations.

**Goals**

- Based on the existing issue, the e-commerce company aims to develop a system that can predict whether customers will churn or not in the future.

- The company also wants to identify the features or factors that influence whether customers will churn or not.

**Analytic Approach**

The analytical approach used involves analyzing customer behavior patterns on the e-commerce platform, which can be used to classify customers predicted to churn and not churn. From this approach, a classification model can be created to predict whether customers will churn or not in the future.

##### **False Positives (FP)**
Customer is predicted to churn, but in reality, they do not churn (stay).

**Negative Impact**
- Opportunity Cost: Additional costs incurred to retain customers who are actually not at risk of churning. The allocation of funds could have been used for acquiring new business opportunities, such as acquiring new customers, expanding partnerships with retail stores, and others.

**Positive Impact**
- Increase in Retention Rate: By providing benefits to customers, loyalty to the e-commerce company increases. This helps mitigate the risk of customers churning or switching to competitors.

#####  **False Negatives (FN)**
Customer is predicted to stay (not churn), but in reality, they churn.

**Negative Impact**
- Revenue Loss: Missing the opportunity to retain customers who are actually at risk of churning. This leads to potential revenue loss from customers who churned and could have continued making purchases if retained.
  
- Cost of Reacquisition: Acquiring new customers to replace those who churned generally incurs higher costs compared to retaining existing customers.
  
- Long-Term Impact: If the customer churn rate is not reduced promptly, it can lead to a decrease in overall customer base, resulting in declining revenue and profits for the company. In the long term, this may impact the company's ability to sustain its business operations and compete effectively against competitors.

Therefore, e-commerce companies need to mitigate the risk of customer churn. Apart from ensuring long-term business sustainability, the cost of customer acquisition is significantly higher than the cost of customer retention.

![CAC vs CRC](https://blog.nextsale.io/wp-content/uploads/2020/05/Image_2.png)

**Churn Risk Mitigation**

The comparison between **customer retention cost : customer acquisition cost = 1 : 5**, where **customer acquisition cost** is **5 - 7 times** higher than **customer retention cost**.

Here are other supporting reasons why e-commerce companies should focus on customer retention rather than customer acquisition:

**a. Higher Return on Investment (ROI)**

Efforts to retain customers typically yield a higher Return on Investment (ROI) compared to efforts to acquire new customers. Research shows that increasing customer retention rates by just 5% can increase company profits by 25% to 95%. This is because retaining existing customers requires fewer resources and costs compared to acquiring new customers, making it a more cost-effective long-term strategy.

**b. Increasing Customer Lifetime Value (CLV)**

Loyal customers tend to make more transactions, add-on purchases, and recommend others to use the same e-commerce service (indirectly assisting the company's marketing efforts, whether through social media or word of mouth). A steady stream of revenue from loyal customers has the potential to boost the revenue and profitability of the e-commerce company in the long term.

**c. Reducing Marketing Costs**

Acquiring new customers requires significant investment in marketing campaigns and advertising to increase awareness and attract potential customers. In contrast, retaining existing customers by maintaining ongoing relationships can reduce the marketing costs incurred.

**d. Cross-Selling Opportunities**

Satisfied existing customers are more receptive to additional sales efforts. By leveraging customer data and behavior, e-commerce companies can identify opportunities to recommend other products that align with customer preferences. This not only has the potential to increase the company's revenue but also strengthens overall customer relationships and enhances customer satisfaction with the e-commerce services provided.

- The three columns (`Tenure`, `WarehouseToHome`, `HourSpendOnApp`, `OrderAmountHikeFromlastYear`, `CouponUsed`, `OrderCount`, `DaySinceLastOrder`) with missing values have data distributions that are not normal or skewed to the right.
- Therefore, missing values in each column can be filled with the median value.
- They are not filled using techniques like iterative imputer or KNN imputer because the values in each column are integers. Iterative imputer or KNN imputer techniques could potentially produce decimal numbers, which would not be suitable in this case.

```python
list_mean = []
list_median = []
list_mode = []

for i in missing_values_columns: 
    mean = df[i].mean()
    median = df[i].median()
    mode = df[i].mode()[0]

    list_mean.append(mean)
    list_median.append(median)
    list_mode.append(mode)
```

```python
df_mean_median_mode = pd.DataFrame()
df_mean_median_mode["Missing Values Column"] = missing_values_columns
df_mean_median_mode["Mean Value"] = list_mean
df_mean_median_mode["Median Value"] = list_median
df_mean_median_mode["Mode Value"] = list_mode

df_mean_median_mode
```

- The four columns (`HourSpendOnApp`, `CouponUsed`, `OrderCount`, `DaySinceLastOrder`) have the same median and mode values.
- The three columns (`Tenure`, `WarehouseToHome`, `OrderAmountHikeFromlastYear`,) have different median and mode values. 
- Based on the histograms above, we will use median values to fill the missing values because the data are not normally distributed. 

```python
# Number of rows with median values in columns containing missing values
# Before filling

for i in missing_values_columns:
    print(i)
    print(len(df[df[i] == df[i].median()]))
    print()
```

```python
# Fill missing values

for i in missing_values_columns:
    df[i].fillna(df[i].median(), inplace = True)
```

```python
# Number of rows with median values in columns containing missing values
# After filling 

for i in missing_values_columns:
    print(i)
    print(len(df[df[i] == df[i].median()]))
    print()
```

```python
def churn_rate(feature):
    pivot_table = pd.pivot_table(data = df, columns = "Churn", index = feature, aggfunc = "count", values = "CustomerID")
    pivot_table["Total"] = pivot_table.sum(axis = 1)
    pivot_table["Churn Rate"] = pivot_table[1] / pivot_table["Total"]
    return pivot_table
```

```python
for i in object_columns:
    print(churn_rate(i).sort_values(by = "Churn Rate", ascending = False))
    print()
```

## **8. Cross Validation**

```python
# Cross validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold, KFold

# Algorithm
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm

# Metric
from sklearn.metrics import recall_score
```

```python
# Define algorithm
logreg = LogisticRegression(random_state = 0)
knn = KNeighborsClassifier()
dectree = DecisionTreeClassifier(random_state = 0)
rf = RandomForestClassifier(random_state = 0)
gradboost = GradientBoostingClassifier(random_state = 0)
xgb = XGBClassifier()
gnb = GaussianNB()
svm = svm.SVC(kernel = "linear", random_state = 0)
```

```python
list_algo = [logreg, knn, dectree, rf, gradboost, xgb, gnb, svm]
list_all_acc = []
list_mean_acc = []
list_std_acc = []

for model in list_algo:

    # Pipeline
    pipe_prep_model = Pipeline([
        ("preprocessing", transformer),
        ("modeling", model),
    ])

    # Define KFold 
    fold = StratifiedKFold(n_splits = 5)

    # Cross validation score
    cross_val = cross_val_score(estimator = pipe_prep_model,    
                                X =  X_train,
                                y = y_train,
                                cv = fold,                      
                                scoring = "recall") 
    
    # List recall score & std
    list_all_acc.append(cross_val.round(2))
    list_mean_acc.append(cross_val.mean())
    list_std_acc.append(cross_val.std())
```

**Explanation**
In this code, we applied various machine learning models to evaluate their performance using cross-validation, focusing on recall as the metric. The key steps and findings are summarized below:

1. Model Evaluation: We evaluated a range of algorithms, including Logistic Regression (logreg), K-Nearest Neighbors (knn), Decision Tree (dectree), Random Forest (rf), Gradient Boosting (gradboost), XGBoost (xgb), Gaussian Naive Bayes (gnb), and Support Vector Machine (svm).

2. Pipeline Setup: For each model, we built a pipeline that combined preprocessing steps (transformer) and the specific algorithm (model). This helped in standardizing the process.

3. Cross-Validation: We used a 5-fold stratified cross-validation strategy (StratifiedKFold) to ensure balanced class distributions across folds.

4. Recall Scores: For each model, we computed the recall scores across the 5 cross-validation folds. The recall is important for our use case, likely because of class imbalance or a focus on minimizing false negatives (e.g., in a churn prediction problem, identifying all churners is critical).

5. Results:
    - Raw Recall Scores: We stored the individual recall scores for each fold in list_all_acc, rounded to 2 decimal places.
    - Mean Recall: We calculated the mean recall score for each model and saved it in  list_mean_acc, which helps compare the models’ average performance.
    - Standard Deviation: The standard deviation of recall scores was stored in list_std_acc, which gives an idea of the consistency of each model's performance.

```python
# DataFrame Cross Validation

df_cv = pd.DataFrame()
df_cv["Algorithm"] = list_algo
df_cv["All Score"] = list_all_acc
df_cv["Mean Score"] = list_mean_acc
df_cv["STD Score"] = list_std_acc

df_cv.sort_values("Mean Score", ascending = False)
```

## **10. Hyperparameter Tuning**

Hyperparameter tuning is a crucial step in optimizing machine learning models to achieve the best possible performance. Unlike model parameters that are learned from the training data (like weights in neural networks), hyperparameters are set before the training process and control the learning process itself. Examples include the learning rate, number of trees in a random forest, or the number of layers in a neural network.

### **Why Hyperparameter Tuning Is Important:**

**1. Performance Optimization:** Different hyperparameters can significantly impact a model's performance. Finding the optimal set of hyperparameters can lead to better accuracy, recall, precision, or other relevant metrics.

**2. Model Generalization:** Properly tuned hyperparameters help the model generalize better to unseen data, avoiding overfitting (where the model performs well on training data but poorly on new data) or underfitting (where the model performs poorly on both training and testing data).

### **How Hyperparameter Tuning Works:**
1. **Define the Hyperparameters to Tune:** Identify which hyperparameters are crucial for your model and set their possible values or ranges. For example, in XGBoost, common hyperparameters to tune include learning_rate, max_depth, and n_estimators.

2. **Choose a Tuning Method:**

    - Grid Search: This method involves specifying a grid of hyperparameter values and evaluating the model performance for each combination. It’s exhaustive but can be computationally expensive.

    - Random Search: Instead of trying all possible combinations, random search samples a subset of the hyperparameter space. This method can be more efficient and still yield good results.

**3. Evaluate Performance:** For each combination of hyperparameters, the model is trained and evaluated using cross-validation or a validation set. The performance metrics (such as recall) are then compared to find the best set of hyperparameters.

**4. Select the Best Hyperparameters:** The combination that provides the best performance metrics is selected. This set is then used to train the final model on the entire training dataset.

Based on the results of hyperparameter tuning using GridSearchCV, the best XGBClassifier model was found with the following parameters:


1. *colsample_bytree* (modeling__colsample_bytree: 1.0):

    - Definition: This parameter represents the fraction of features (columns) to be randomly sampled for each tree. It controls the proportion of features that will be used to train each individual tree in the ensemble.

    - Value (1.0): This means that all features (100%) are used for every tree. Reducing this can help prevent overfitting but can also reduce the model's predictive power if set too low.

2. *gamma* (modeling__gamma: 0):

    - Definition: The minimum loss reduction required to make a further partition on a leaf node of the tree. It controls whether a node will split based on the expected improvement in the objective function.

    - Value (0): This means there’s no constraint on making splits. Higher values would make the algorithm more conservative, preventing it from creating nodes unless a significant improvement is observed.

3. *learning_rate* (modeling__learning_rate: 0.2):

    - Definition: This controls the step size at each boosting iteration. Lower values make the model learn more slowly but often result in better generalization, as it reduces the risk of overfitting.

    - Value (0.2): A learning rate of 0.2 means the model updates the weights of the trees relatively quickly. Smaller learning rates would need more boosting rounds (n_estimators) to converge.

4. *max_depth* (modeling__max_depth: 15):

    - Definition: The maximum depth of a tree. Increasing the depth allows the model to learn more complex patterns, but too deep trees may lead to overfitting.

    - Value (15): This allows the trees to grow quite deep, meaning the model can capture intricate relationships in the data. However, deeper trees may lead to overfitting, especially on smaller datasets.

5. *min_child_weight* (modeling__min_child_weight: 1):

    - Definition: This controls the minimum sum of instance weights (Hessian) required in a child. It is a regularization parameter used to control overfitting. If the sum of weights in a node is less than min_child_weight, the node will not be further divided.

    - Value (1): A low value like 1 allows the model to create children nodes with fewer samples, making the model more sensitive to small data patterns, potentially increasing the risk of overfitting.

6. *n_estimators* (modeling__n_estimators: 300):

    - Definition: The number of trees (boosting rounds) in the ensemble. More trees can improve model performance, but it also increases training time.

    - Value (300): This means the model will build 300 trees. While more trees can increase the accuracy, the learning rate and early stopping techniques often control when the model has learned enough.

7. *subsample* (modeling__subsample: 0.8):

    - Definition: The fraction of the training data that is randomly sampled to grow each tree. It controls the trade-off between bias and variance.

    - Value (0.8): This means that 80% of the training data is used to train each tree. Lowering this value can prevent overfitting by introducing more variability between trees, but setting it too low can lead to underfitting.

*Summary of Effects:*
- colsample_bytree (1.0): Uses all features for each tree.
- gamma (0): No minimum threshold for making tree splits.
- learning_rate (0.2): Moderately fast learning, which means fewer boosting rounds are needed compared to lower values.
- max_depth (15): Very deep trees to capture complex relationships.
- min_child_weight (1): More sensitive to smaller patterns, allowing splits even with few samples.
- n_estimators (300): The model builds 300 trees, benefiting from the relatively high learning rate.
- subsample (0.8): Each tree is trained on 80% of the data, helping to reduce overfitting.
